# Weather Data Analysis

```python
# Combined Weather Data Analysis
import pandas as pd
df = pd.read_csv(r"P:\Weather-Data-Analysis\datasets\dataset.csv")
```

```python
# Data overview
print(df.head())
print(df.shape)
print(df.index)
print(df.columns)
print(df.dtypes)
```

```python
# Unique values and counts
print(df['Weather'].unique())
print(df.nunique())
print(df.count())
print(df['Weather'].value_counts())
print(df.info())
```

```python
# Wind speed analysis
print(df['Wind Speed_km/h'].nunique())
print(df['Wind Speed_km/h'].unique())
```

```python
# Filtering and grouping
print(df[df.Weather == 'Clear'])
print(df.groupby('Weather').get_group('Clear'))
print(df[df['Wind Speed_km/h'] == 4])
```

```python
# Missing values
print(df.isnull().sum())
print(df.notnull().sum())
```

```python
# Rename column
df.rename(columns = {'Weather' : 'Weather Condition'}, inplace = True)
print(df.head())
```

```python
# Statistics
print(df.Visibility_km.mean())
print(df.Press_kPa.std())
print(df['Rel Hum_%'].var())
```

```python
# Weather Condition analysis
print(df['Weather Condition'].value_counts())
print(df[df['Weather Condition'] == 'Snow'])
print(df[df['Weather Condition'].str.contains('Snow')].tail(50))
```

```python
### Advanced filtering
print(df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)])
print(df.groupby('Weather Condition').mean(numeric_only=True))
print(df.groupby('Weather Condition').min())
print(df.groupby('Weather Condition').max())
print(df[df['Weather Condition'] == 'Fog'])
print(df[(df['Weather Condition'] == 'Clear') | (df['Visibility_km'] > 40)].tail(50))
print(df[(df['Weather Condition'] == 'Clear') & (df['Rel Hum_%'] > 50)|(df['Visibility_km'] > 40)])
```

```python
# Plot: Average Temperature by Weather Condition
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_plot = pd.read_csv(r"P:\Weather-Data-Analysis\datasets\dataset.csv")
df_plot.rename(columns = {'Weather' : 'Weather Condition'}, inplace = True)
avg_temp = df_plot.groupby('Weather Condition')['Temp_C'].mean().sort_values()

plt.figure(figsize=(12,6))
sns.barplot(x=avg_temp.index, y=avg_temp.values)
plt.xticks(rotation=90)
plt.xlabel('Weather Condition')
plt.ylabel('Average Temperature (C)')
plt.title('Average Temperature by Weather Condition')
plt.tight_layout()
plt.show()
```
